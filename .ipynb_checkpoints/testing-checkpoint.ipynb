```python
def getMeanOfTwoGaussians(mu1,mu2,var1,var2):
    return float(mu1*var2 + mu2*var1)/float(var1 + var2)
```

```python
def getVarOfTwoGaussians(var1, var2):
    #print (var1, var2)
    return float(var1*var2)/float(var1 + var2)
```

```python
def getPosteriorMeanAndStd(N, m, priorMean =0, priorStd = 100):
    mean = 0
    var = math.pow(priorStd, 2)
    
    for i in range(N):
        mean = getMeanOfTwoGaussians(mean, m, var, 1)
        #print mean
        var = getVarOfTwoGaussians(var, 1)
        #print var
    return mean, math.sqrt(var)
```

```python
getPosteriorMeanAndStd(N, m)
```

```python
def runExperiment(N, m):
    posteriorMean, posteriorStd = getPosteriorMeanAndStd(N, m)
    isWithinOneStd = np.abs(m - posteriorMean) < 1
    return isWithinOneStd
```

```python
runExperiment(10, -100001)
```
